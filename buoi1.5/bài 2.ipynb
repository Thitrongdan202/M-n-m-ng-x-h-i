```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([(1, 3), (2, 4), (2, 6), (3, 1), (3, 2), (4, 5), (5, 1), (5, 4), (6, 1), (6, 2), (6, 3), (6, 4)])
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)  # Position nodes using spring layout
nx.draw_networkx(G, pos, node_size=600, node_color='skyblue', font_size=14, font_color='black', arrows=True)

plt.title("Directed Graph Visualization")
plt.axis('off')  # Turn off axis
plt.show()
```

```python
G1 = nx.DiGraph()
G1.add_edges_from([(1,2),(1,3),(1,4),(2,5),(2,7),(3,1),
                   (3,2),(3,3),(4,6),(5,1),(5,4),(6,1),(6,2),(6,4),(7,6),(7,7)])
```

```python
plt.axis('off')
nx.draw_networkx(G1, pos= nx.spring_layout(G1, seed = 0), node_size=600, cmap= 'coolwarm', font_size=14, font_color='white')
```
